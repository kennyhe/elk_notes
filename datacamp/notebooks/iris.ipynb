```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/iris.data', names=['sepal_len', 'sepal_width', 'petal_len', 'petal_width', 'class'])
```

```python
df.iloc[:3, [1, 3]]
```

```python
plt.style.use('ggplot')
%matplotlib inline

fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['petal_width'], color='green')
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Width', fontsize=12)
plt.title('Iris Petal Width', fontsize=14, y=1.01)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(6,4))

ax[0][0].hist(df['petal_width'], color='green')
ax[0][0].set_ylabel('Count', fontsize=12)
ax[0][0].set_xlabel('Width', fontsize=12)
ax[0][0].set_title('Iris Petal Width', fontsize=14, y=1.01)

ax[0][1].hist(df['petal_len'], color='red')
ax[0][1].set_ylabel('Count', fontsize=12)
ax[0][1].set_xlabel('Width', fontsize=12)
ax[0][1].set_title('Iris Petal Length', fontsize=14, y=1.01)

ax[1][0].hist(df['sepal_width'], color='blue')
ax[1][0].set_ylabel('Count', fontsize=12)
ax[1][0].set_xlabel('Width', fontsize=12)
ax[1][0].set_title('Iris Sepal Width', fontsize=14, y=1.01)

ax[1][1].hist(df['sepal_len'], color='yellow')
ax[1][1].set_ylabel('Count', fontsize=12)
ax[1][1].set_xlabel('Width', fontsize=12)
ax[1][1].set_title('Iris Sepal Length', fontsize=14, y=1.01)

plt.tight_layout()
```

```python
flg, ax=plt.subplots(figsize=(6,6))
ax.scatter(df['petal_width'], df['petal_len'], color='green')
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatter plot')
```

```python
flg, ax=plt.subplots(figsize=(6,6))
ax.plot(df['petal_len'], color='blue')
ax.set_xlabel('Specimen Number')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Length plot')
```

```python
flg, ax=plt.subplots(figsize=(6,10))
bar_width = .8
labels = [x for x in df.columns if 'len' in x or 'width' in x]
ver_y = [df[df['class'] == 'Iris-versicolor'][x].mean() for x in labels]
vir_y = [df[df['class'] == 'Iris-virginica'][x].mean() for x in labels]
set_y = [df[df['class'] == 'Iris-setosa'][x].mean() for x in labels]
x = np.arange(len(labels))
ax.bar(x, vir_y, bar_width, bottom=set_y, color='blue')
ax.bar(x, set_y, bar_width, bottom=ver_y, color='red')
ax.bar(x, ver_y, bar_width, color='green')
ax.set_xticks(x + (bar_width/2))
ax.set_xticklabels(labels)
ax.set_title('Mean Feature Measurement By Class')
ax.legend(['Virginica', 'Setosa', 'Versicolor'])
```
